```python
import requests
from captcha.image import ImageCaptcha
import matplotlib.pyplot as plt
import random
import string
```

```python
REST_API_URL = "http://localhost:5000/predict"
characters = string.digits + string.ascii_uppercase
```

## Randomly generate a captcha image

```python
#random_str = ''.join([random.choice(characters) for j in range(4)])
random_str = 'O0O0'
generator = ImageCaptcha(width=170, height=80)
img = generator.generate_image(random_str)

generator.write(random_str, 'captcha.png')
plt.imshow(img)
plt.title(random_str)
plt.axis('off')
```

## Send the captcha to service, return the predicted text

```python
img = open('captcha.png', "rb").read()
pred = requests.post(REST_API_URL, files={"image": img}).json()
print("The text of requested captcha: {}".format(pred))
```
